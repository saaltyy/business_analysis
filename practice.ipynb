```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Invistico_Airline.csv', on_bad_lines='skip')
data
```

```python
data.drop(['Arrival Delay in Minutes'], axis = 1)
```

```python
for column_name in data.columns:
    if data[column_name].dtype in ['int64', 'float64']:
        plt.hist(data[column_name], bins=20)
        plt.title(f'Histogram of {column_name}')
        plt.xlabel(column_name)
        plt.ylabel('Frequency')
        plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Matrix Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(data=data, x="Class", y="Departure Delay in Minutes")
plt.title("Class vs. Departure Delay in Minutes")
plt.xlabel("Class")
plt.ylabel("Departure Delay in Minutes")

plt.show()
```

```python
poddata = data.groupby(['Customer Type', 'satisfaction']).size()
poddata = poddata.unstack()
poddata.plot(kind = 'bar', figsize=(6,4), width = 1, stacked=True, grid = True, alpha = 0.8, ec = 'black');
plt.title('Types of vehicle and the fuel they use', fontsize = "10")
plt.xlabel('')
plt.show();
```

```python
for column_name in data.columns:
    if data[column_name].dtype == 'object':
        plt.figure(figsize=(4, 2))
        sns.countplot(x=column_name, data=data)
        plt.title(f'Count of {column_name}')
        plt.xticks(rotation=90)
        plt.show()
```
